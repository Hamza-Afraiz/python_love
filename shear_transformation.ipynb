```python
#showing shear transformation
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as tck
fig = plt.figure(figsize=(12,12))
# Compute the x and y coordinates for points on a sine curve
x = np.arange(-15, 15, 1)

y = np.arange(-15, 15, 1)
plt.xlim(-15,15)
plt.ylim(-15,15)
ax = plt.gca()
ax.yaxis.set_major_locator(tck.MultipleLocator(1))
ax.yaxis.set_minor_locator(tck.MultipleLocator(1))
ax.xaxis.set_major_locator(tck.MultipleLocator(1))
ax.xaxis.set_minor_locator(tck.MultipleLocator(1))
ax.grid()


arr = np.stack((x, y), axis=1)

d = np.array([[1,1],[0,1]])
#print("d is ",d)

#print(arr)
ar=np.array([[0],[0]])
xy=np.array([[0],[0]])
xyz=np.array([[0],[0]])
#print(ar.shape)


q=(arr.reshape(30,2,1))
#print(q)
#print(q.shape)

for k in x:
    for o in y: 
     #print("k shape is ",k.shape)
     #print("y  is ",o)
     #print("x is ",x[0])
     xy=np.append(xy,[[k],[o]],axis=1)
     #print("k  o is " ,[[k],[o]])
     dd=d.dot([[k],[o]])
     #print("dd is ",dd)
    
  
     xyz=np.append(xyz,dd,axis=1)
     dd=np.array([[0],[0]])
    
     #point=d.dot(x);
     #print("after point is ",point)

asd=xyz[0].ravel()
asf=xyz[1].ravel()
#print(asd)

plt.plot(asd,asf,'bo')
plt.plot(asd,asf)
plt.show()
#print(xyz.shape)



#print(xyz[0][1])
#plt.plot(xyz[0],xyz[1])
```
